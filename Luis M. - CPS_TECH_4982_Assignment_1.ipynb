````python
# Q7-1

moonImg = cv2.imread('blurry-moon.tif', cv2.IMREAD_GRAYSCALE)

print("Minimum pixel value:", np.min(moonImg))
print("Maximum pixel value:", np.max(moonImg))
````

````python
# Q7-2

height = moonImg.shape[0]
width = moonImg.shape[1]

center_pixel = moonImg[height//2, width//2]
print("Center pixel value:", center_pixel)

upperLeftCorner = moonImg[(height//2)-1, (width//2)-1]
print("Upper left corner pixel value:", upperLeftCorner)
up = moonImg[(height//2)-1, width//2]
print("Upper pixel value:", up)
upperRightCorner = moonImg[(height//2)-1, (width//2)+1]
print("Upper right corner pixel value:", upperRightCorner)

right = moonImg[height//2, (width//2)+1]
print("Right pixel value:", right)
left = moonImg[height//2, (width//2)-1]
print("Left pixel value:", left)

lowerLeftCorner = moonImg[(height//2)+1, (width//2)-1]
print("Lower left corner pixel value:", lowerLeftCorner)
down = moonImg[(height//2)+1, width//2]
print("Lower pixel value:", down)
lowerRightCorner = moonImg[(height//2)+1, (width//2)+1]
print("Lower right corner pixel value:", lowerRightCorner)
````

````python
# Q7-3

manipulated_moonImg = moonImg.copy()

manipulated_moonImg = np.where((manipulated_moonImg >= 100) & (manipulated_moonImg <= 200), 255, 0)

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(moonImg, cmap = "gray")
plt.axis('off')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(manipulated_moonImg, cmap = "gray")
plt.axis('off')
plt.title('Manipulated Image')

plt.show()
````
